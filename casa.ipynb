```python
import pandas as pd

bd = pd.read_csv('competitor.csv')
```

```python
dim_competitors = bd[[
            'competitor.id',
            'competitor.sum_points',
            "competitor.place",
            "competitor.place_prev",
            "competitor.id_person",
            "competitor.family_name",
            "competitor.given_name",
            "competitor.gender",
            "competitor.timestamp_version",
            "competitor.id_country",
            "competitor.id_weight",]].drop_duplicates(subset=['competitor.id'])
dim_competitors.head()
```

```python
dim_competitors.shape
```

```python
dim_weight_category = bd[['competitor.id_weight', 'competitor.weight_name', 'competitor.gender']].drop_duplicates().sort_values(['competitor.id_weight']).reset_index(drop=True)
dim_weight_category
```

```python
dim_country = bd[['competitor.id_country', 'competitor.country_name', 'competitor.country_ioc_code', 'competitor.id_continent']].drop_duplicates().sort_values('competitor.id_country').reset_index(drop=True)
dim_country
```

```python
dim_continent = dim_country[['competitor.country_name','competitor.country_ioc_code','competitor.id_continent']].drop_duplicates('competitor.id_continent').sort_values('competitor.id_continent').assign(continent_name =  ['ijf', 'Africa', 'Asia', 'Europe', 'Oceania', 'America'])
dim_continent
```

```python
dim_continent = dim_continent[['competitor.id_continent','continent_name']].reset_index(drop=True)
dim_continent
```

```python
dim_competition = bd[['competition.id', 'competition.name', 'competition.date_from']].drop_duplicates().sort_values('competition.date_from').reset_index(drop=True)
dim_competition
```

```python
fact_tournament_results = bd[['points', 'place', 'long_place', 'competition.id', 'competitor.id']]
fact_tournament_results
```
